```python
#Como en cualquier base de datos, necesitamos distintas herramientas para poder llevar a cabo nuestro análisis, las librerías principales que utilizaremos son las siguientes
import pandas as pd
import numpy as np
```

```python
#A continuación lo que hacemos es crear una variable, en este caso df, y le decimos a programa que descargue el cvs con el nombre "Ventas_totales_supermercados.csv" que es el que utilizaremos durante esta actividad. 
df = pd.read_csv("Ventas_totales_supermercados.csv", index_col=0)
df
```

```python
# Filtro UNO: 
filtro1 =df[df["ventas_precios_corrientes"] == "ventas_precios_constantes"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
```

```python
#Filtro DOS: Filtro por filas, vamos a utilizar el siguiente código ya que las filas que vamos a seleccionar para este filtro no son consecutivas. 
#filtro2= df.iloc[300:302,: ]
filtro2=df.iloc[ [10, 20, 30, 40, 50,60] , : ]  #filas no consecutivas 
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro2.csv")
```

```python
#Filtro TRES:Filtro por columnas, al igual que el filtro de arriba, utilizamos este código ya que nuestras columnas no son consecutivas. 
filtro3=df.iloc[ : , 4:7]
#filtro3=df.iloc[ : , [6, 12, 18,23]]  #columnas no consecutivas 
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro3.csv")
```

```python
#Filtro CUATRO: filtro de filas vs columnas, este filtro como su nombre lo dice, seleccionamos de manera manual y más específica las columnas y filas que queremos para nuestro análisis. 
filtro4=df.loc[["2017-04-01","2022-03-01"], ["ventas_precios_corrientes"]]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("Filtro4.csv")
```

```python
#Filtro CINCO: filtro por cabecera, este filtro nos ayuda a tener una idea de nuestra base de datos, nosotros seleccionamos que queríamos las primeras cinco filas, este número puede variar. 
filtro5=df.head()
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("Filtro5.csv")
```

```python
#Filtros SEIS: Filtros básicos para variables cuantitativas 
#Filtro por comparacion, este filtro nos ayuda para obtener información más específica con datos numéricos. 
filtro6=df[df["ventas_precios_corrientes"] > 27969.0]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro6.csv")
```

```python
#FILTROS BASICOS CON CONETORES LOGICOS
#Filtro Y
filtro7=df[(df["ventas_precios_corrientes"] > 152202.6) & (df["indice_tiempo"] == "2017-04-01")]
filtro7
```

```python
#Filtro siete correcto
filtro7=df[(df["ventas_precios_corrientes"] > 152202.6) & (df["ventas_totales_medio_pago"] >3.115488e+07)]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro7.csv")
```

```python
#Filtro o este filtro nos ayuda a comparra dos variables con filtros específicos 
filtro8=df[(df["salon_ventas"] > 1.496976e+08) | (df["canales_on_line"] == 4.819480e+06	)]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro8.csv")
```

```python
#Filtro not
filtro9=df[~(df["salon_ventas"] >= 1.496976e+08) & (df["canales_on_line"] == 4.819480e+06)]
filtro9
```
